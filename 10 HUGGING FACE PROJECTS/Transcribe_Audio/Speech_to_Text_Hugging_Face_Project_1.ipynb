```python
import transformers
import librosa
import torch
import numpy as np
from transformers import pipeline
import IPython.display as display
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
!wget https://github.com/tanviredu/AILEARNING/raw/refs/heads/master/10%20HUGGING%20FACE%20PROJECTS/BreakingBad.mp4
```

```python
audio,sampling_rate = librosa.load('BreakingBad.mp4',sr=16000)
```

```python
display.Audio(audio,rate=sampling_rate)
```

```python
input_values = tokenizer(audio, return_tensors="pt").input_values
input_values
```

```python
non_normalizedd_logits = model(input_values).logits
non_normalizedd_logits
```

```python
predicted_ids = torch.argmax(non_normalizedd_logits, dim=-1)
```

```python
transcription = tokenizer.decode(predicted_ids[0])
print(transcription)
```
